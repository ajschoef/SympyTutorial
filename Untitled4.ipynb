As an example, let's use sympy to solve a cubic equation, the van der Waals equation


```python
import sympy

# First, let's designate the variables we want to be symbolic 
# and define those for which we  have values.

Vm = sympy.symbols('Vm')
P = 1.48
R = 0.082057
T = 250
a = 6.2
b = 0.0542

# Next, let's write down the van der Waals equation, using our variables 

vdw = P*(Vm**3) - b*P*(Vm**2) - R*T*(Vm**2) + a*Vm - a*b

# Finally, we can evaluate the solutions of the van der Waals equation

sympy.roots(vdw, Vm)
```

```python
# to show vdw vs. perfect
import matplotlib.pyplot as plt
import numpy as np

V = np.linspace(0.048,0.5,100)
R = 0.082057
Tlist=[205]
# Carbon dioxide
a = 3.610
b = 0.0419
def vdw(V, R, T, a, b):
    return ((R*T)/(V - b)) - (a/(V**2))

for T in Tlist:
    p = vdw(V, R, T, a, b)
    plt.plot(V, p)
plt.plot(V, R*205/V)
plt.show()
    
```
